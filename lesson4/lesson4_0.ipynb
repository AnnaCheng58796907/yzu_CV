```python
# 影像操作與運算
# 影像遮罩範例
import cv2
import numpy as np
img1 = np.zeros((200, 300, 3), np.uint8)
img1[:, :, 1] = 255
cv2.imshow("img1", img1)
img2 = np.zeros((200, 300, 3), np.uint8)
img2[:, :, 2] = 255
cv2.imshow("img2", img2)
m = np.zeros((200, 300, 1), np.uint8)
m[50:150, 100:200, :] = 255
cv2.imshow("mask", m)
img3 = cv2.add(img1, img2)
cv2.imshow("img1 + img2", img3)
img4 = cv2.add(img1, img2, mask=m)
cv2.imshow("img1 + img2 + mask", img4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 影像操作與運算
# 重複曝光範例
import cv2
import numpy as np
src1 = np.ones((2, 3), dtype=np.uint8) * 10
src2 = np.ones((2, 3), dtype=np.uint8) * 50
alpha = 1
beta = 0.5
gamma = 5
print(f"src1 = \n {src1}")
print(f"src2 = \n {src2}")
dst = cv2.addWeighted(src1, alpha, src2, beta, gamma)
print(f"dst = \n {dst}")
```

```python
# 影像操作與運算
# 影像的位元運算範例
import cv2
import numpy as np
src1 = np.random.randint(0, 255, (3, 5), dtype=np.uint8)
src2 = np.zeros((3, 5), dtype=np.uint8)
src2[0:2, 0:2] = 255
dst = cv2.bitwise_and(src1, src2)
print(f"src1 = \n {src1}")
print(f"src2 = \n {src2}")
print(f"dst = \n {dst}")
```

```python
# 影像操作與運算
# 使用位元運算進行影像遮罩
import cv2
import numpy as np
src1 = cv2.imread("leclerc_small.png", cv2.IMREAD_GRAYSCALE)
src2 = np.zeros(src1.shape, dtype=np.uint8)
src2[30:260, 70:260] = 255
dst = cv2.bitwise_and(src1, src2)
cv2.imshow("Leclerc", src1)
cv2.imshow("Mask", src2)
cv2.imshow("Result", dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# 影像操作與運算
# 影像加密與解密的範例
import cv2
import numpy as np
src = cv2.imread("leclerc_small.png")
key = np.random.randint(0, 256, src.shape, np.uint8)
print(src.shape)
cv2.imshow("Leclerc", src)
cv2.imshow("Key", key)
img_encry = cv2.bitwise_xor(src, key)           # 加密
img_decry = cv2.bitwise_xor(key, img_encry)     # 解密
cv2.imshow("Encrytion", img_encry)
cv2.imshow("Decrytion", img_decry)
cv2.waitKey()
cv2.destroyAllWindows()
```
